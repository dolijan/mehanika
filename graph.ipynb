```python
import numpy as np
import matplotlib.pyplot as plt

m = 70.0  
g = 9.8      
k = 150.0       
d = 20.0      
cv = 60    
dt = 0.01      
t_max = 60.0    

t0 = 0.0
x0 = 0.0       
v0 = 0.0    

vremena = [t0]
pozicije = [x0]
brzine = [v0]

t = t0
x = x0
v = v0

while t < t_max:

    if x <= d:
        a = g
    else:
        a = g - (k * (x - d)) / m - (cv * v) / m
    
    v_nova = v + a * dt
    x_nova = x + v_nova * dt
    t_novo = t + dt

    vremena.append(t_novo)
    pozicije.append(x_nova)
    brzine.append(v_nova)

    t = t_novo
    x = x_nova
    v = v_nova

plt.figure(figsize=(10,6))
plt.plot(vremena, pozicije, label='Pozicija (m)')
plt.axhline(d, color='red', linestyle='--', label='Prirodna dužina trake (d)')
plt.xlabel('Vreme (s)')
plt.ylabel('Pozicija (m)')
plt.title('Simulacija bungee skoka')
plt.legend()
plt.grid(True)
plt.show()
```
