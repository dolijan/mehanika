```python
import numpy as np
import matplotlib.pyplot as plt

# Parametri
m = 70.0        # masa (kg)
g = 9.8        # gravitacija (m/s^2)
k = 150.0        # konstanta opruge (N/m)
d = 20.0        # prirodna dužina elastične trake (m)
cv = 60.0        # koeficijent prigušenja (kg/s)
dt = 0.01       # vremenski korak (s)
t_max = 60.0    # ukupno vreme simulacije (s)

# Početni uslovi
t0 = 0.0
x0 = 0.0        # početna pozicija (tačka kačenja)
v0 = 0.0        # početna brzina

# Liste za čuvanje rezultata
vremena = [t0]
pozicije = [x0]
brzine = [v0]

t = t0
x = x0
v = v0

while t < t_max:
    # Izračunavanje ubrzanja a(t)
    if x <= d:
        a = g
    else:
        a = g - (k * (x - d)) / m - (cv * v) / m
    
    # Ažuriranje brzine i pozicije
    v_nova = v + a * dt
    x_nova = x + v_nova * dt
    t_novo = t + dt

    # Čuvanje rezultata
    vremena.append(t_novo)
    pozicije.append(x_nova)
    brzine.append(v_nova)

    # Priprema za sledeću iteraciju
    t = t_novo
    x = x_nova
    v = v_nova

# Crtanje grafikona pozicije u zavisnosti od vremena
plt.figure(figsize=(10,6))
plt.plot(vremena, pozicije, label='Pozicija (m)')
plt.axhline(d, color='red', linestyle='--', label='Prirodna dužina trake (d)')
plt.xlabel('Vreme (s)')
plt.ylabel('Pozicija (m)')
plt.title('Simulacija bungee skoka')
plt.legend()
plt.grid(True)
plt.show()
```
